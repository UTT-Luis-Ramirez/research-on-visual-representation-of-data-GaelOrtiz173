## Diagrama de Tallo y Hoja

### ¿Para que sirven? y ¿Como son?

#### Es una herramienta grafica utilizada en estadistica para presenyat la distribucion de una variable cuantitativa
#### El tallo consiste en las cifras delanteras del numero, mientras que la hoja es el ultimo digito, por ejemplo el numero 45, el 4 seria el tallo y el 5 seria la hoja

### ¿En que contexto se utiliza?

#### Se usa principalmente en el analisis exploratorio de datos y las estadisticas descriptivas. Es util para organizar grandes cantidades de valores numericos de manera que sean visualmente mas faciles de comprender y analizar.

### ¿Como se crea?

### Primero hay que instalar la libreria en la Terminal de Visual con el siguiente comando: pip install stemgraphic
### Despues de instalar la libreria hay que imporyat el modulo y crear el diagrama de tallo y hoja con los datos que desees

```python
import stemgraphic

#Datos a representar
data = [2, 3, 5, 10, 15, 28, 30, 50]

#Crear el diagrama de tallo y hoja
stemgraphic.stem_graphic(data, scale=10)
```

## Diagrama de caja

### ¿Para que sirven? y ¿Como son?

#### Es una herramienta grafica que se utiliza para representar visualmente una serie de datos numericos a traves de sus cuartiles.
#### Se utilizan para comparar la proximidad de otros valores de datos y para identificar posibles valores atipicos.

### ¿En que contexto se utiliza?

#### Ayuda a los analistas e investigadores a comprender rapidamente la distribucion y la estructura de los datos

### ¿Como se crea?

#### Primero hay que instalar la libreria y se instala con el siguiente comando con la terminal: pip install numpy 
#### Despues hay que importar la libreria y generar los datos y el grafico de la caja

```python
import numpy as np
import matplotlib.pyplot as plt

#Generamos datos aleatorios
np.random.seed(5)
x = np.random.normal(0, 1, 200)

#Creamos el grafico de caja
fig, ax = plt.subplots()
ax.boxplot(x)
plt.show()
```

### El siguiente paso es para generar un conjunto de datos aleatorios y luego crea un grafico de caja con esos datos. Puedes crear tambien un grafico en horizontal utilizando el argumento "vert" como "False"

```python
#Grafico de caja horizontal
np.random.seed(5)
x = np.random.normal(0, 1, 200)
fig, ax = plt.subplots()
ax.boxplot(x, vert=False)
plt.show()
```

### Ademas puedes personalizar el grafico de caja añadiendo mas detalles como la media y la desviacion estandar

```python
# Calculamos la media y desviacion estandar
mean = np.mean(x)
std_dev = np.std(x)

# Creamos el boxplot
plt.boxplot(x)
plt.title('Boxplot con Media y Desviacion Estandar')
plt.ylabel('Valor')

# Añadimos la media como un punro rojo
plt.plot(1, mean, 'ro')

# Añadimos la desviacion estandar como barras de error
plt.errorbar(1, mean, yerr=std_dev, fmt='o', color='red')
plt.show()
```
